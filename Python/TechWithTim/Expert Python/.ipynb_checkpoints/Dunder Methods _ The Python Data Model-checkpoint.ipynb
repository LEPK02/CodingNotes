```python
import inspect

class Pet:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self.age = age
        
    def __str__(self):
        return f"You own a {self.species}."
    
class Dog(Pet):
    def __init__(self, name, species, age):
        super().__init__(name, species, age)
        
    def __repr__(self):
        super().__repr__()
        
    def __call__(self, name):
        if self.name is name:
            print("Woof!")
        else:
            print("Silence...")
            
    def __len__(self):
        return len(self.name)
    
dog = Dog("Tim", "dog", 10)
print(dog)
dog("Tim")
print(len(dog))
```
