<p>Importing Matplotlib and Seaborn</p>

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# Makes graphs are embedded in Jupyter notebook instead of displaying as pop-ups
%matplotlib inline
```

<p>Line Graphs</p>

```python
import random

apple_yield = [random.randint(100, 201) for _ in range(6)]
orange_yield = [random.randint(150, 251) for _ in range(6)]
year = range(2010, 2016)

# Seaborn styling
sns.set_style("whitegrid")

# Matplotlib styling
matplotlib.rcParams["font.size"] = 12
matplotlib.rcParams["font.sans-serif"] = "Arial"

# Set figure size
plt.figure(figsize=(5, 3));

# Plot graph
plt.plot(year, apple_yield, marker="o", color="red");
plt.plot(year, orange_yield, marker="x", color="orange");

# Label x and y axes
plt.xlabel("Year")
plt.ylabel("Yield/kg");

# Set title and legend
plt.title("Crop Yields on Farm")
plt.legend(["Apples", "Oranges"]);
```

<p>Histogram</p>

```python
import numpy as np

flowers_df = sns.load_dataset("iris")
# flowers_df.describe()
setosa_df = flowers_df[flowers_df.species == "setosa"]
versicolor_df = flowers_df[flowers_df.species == "versicolor"]
virginica_df = flowers_df[flowers_df.species == "virginica"]

# Creates array starting from 2 (inclusive) to 5 (exclusive), step = 0.25
bin_list = np.arange(2, 5, 0.25)

# Plot histogram
plt.figure();
plt.hist(setosa_df.sepal_width, bins=bin_list, alpha=0.5);
plt.hist(versicolor_df.sepal_width, bins=bin_list, alpha=0.5);


# # Plot stacked histogram
# plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
#          bins=bin_list,
#          stacked=True);

# Modify parameters
plt.title("Distribution of Sepal Width")
plt.legend(["Setosa", "Versicolor", "Virginica"]);

# Kernel Density Estimate (KDE) plot
plt.figure()
sns.kdeplot(data=setosa_df, x="sepal_width");
sns.kdeplot(data=virginica_df, x="sepal_width");

# Modify parameters
plt.title("Distribution of Sepal Width");
plt.legend(["Setosa", "Virginica"]);
plt.xlabel("Sepal Width");
```

<p>Data Preparation and Cleaning</p>

```python
# Set variables
import pandas as pd

survey_raw_df = pd.read_csv("./stackoverflow-developer-survey-2020/survey_results_public.csv")
schema_raw_df = pd.read_csv("./stackoverflow-developer-survey-2020/survey_results_schema.csv", index_col="Column").QuestionText

selected_columns = [
    # Demographics
    'Country',
    'Age',
    'Gender',
    'EdLevel',
    'UndergradMajor',
    # Programming experience
    'Hobbyist',
    'Age1stCode',
    'YearsCode',
    'YearsCodePro',
    'LanguageWorkedWith',
    'LanguageDesireNextYear',
    'NEWLearn',
    'NEWStuck',
    # Employment
    'Employment',
    'DevType',
    'WorkWeekHrs',
    'JobSat',
    'JobFactors',
    'NEWOvertime',
    'NEWEdImpt'
]

survey_df = survey_raw_df[selected_columns].copy()
schema = schema_raw_df[selected_columns].copy()
```

```python
# View dataframe information
survey_df.tail()
survey_df.shape
survey_df.describe()
survey_df.info()
# survey_df["Age"].unique()
```

```python
import numpy as np

# Convert all strings to floats; if there is an error, NaN will be used as a placeholder value
survey_df["Age1stCode"] = pd.to_numeric(survey_df.Age1stCode, errors="coerce")
survey_df["YearsCode"] = pd.to_numeric(survey_df.Age1stCode, errors="coerce")
survey_df["YearsCodePro"] = pd.to_numeric(survey_df.Age1stCode, errors="coerce")

# Remove "invalid" data (i.e. out of desired range)
#     Makes sure age falls from 10 and 100
survey_df.drop(survey_df[survey_df.Age < 10].index, inplace=True)
survey_df.drop(survey_df[survey_df.Age > 100].index, inplace=True)
#     Ensures working hours are not more than 140 hours (a week has 168 hours)
survey_df.drop(survey_df[survey_df.WorkWeekHrs > 140].index, inplace=True)
#     Removes respondents' answers that accidentally selected more than one gender option (Male, Female, Other)
survey_df.where(~(survey_df.Gender.str.contains(";", na=False)), other=np.nan, inplace=True)
# survey_df.Gender.value_counts()

# survey_df.describe()
# survey_df.sample(10)
```

```python
"""Examining biases in demographics"""
survey_df.Country.nunique()
top_countries = survey_df.Country.value_counts().head(15)

# Checking distribution by country
plt.figure(figsize=(12,6))
plt.title(schema.Country)
plt.pie(x=top_countries, labels=top_countries.index)
plt.show()

# Checking age distribution
plt.figure(figsize=(12,6))
plt.xlabel("Age")
plt.ylabel("Number of Respondents")
plt.title(schema.Age)
plt.hist(x=survey_df.Age, bins=np.arange(10, 100, 5))
plt.show()
```

```python
"""Processing data"""

# Takes input of one column containing lists of values, outputs as dataframe with one column for each possible option
def split_multicolumn(col_series, sep=";"):
#     Converts series to dataframe
    result_df = col_series.to_frame()
    options = []
    
#     Iterate through column
    for idx, val in col_series[col_series.notnull()].iteritems():
#         Iterate through each item
        for option in val.split(sep):
#         Adds new item as a column to new dataframe
            if option not in result_df.columns:
                options.append(option)
                result_df[option] = False
#             Marks value under new dataframe as True
            result_df.at[idx, option] = True
    
    return result_df[options]

dev_type_df = split_multicolumn(survey_df.DevType)
dev_type_totals = dev_type_df.sum().sort_values(ascending=False).head()
print(dev_type_totals)
```

```python
"""Asking and Answering Questions"""
# Which were the most popular programming languages in 2020?
languages_worked_df = split_multicolumn(survey_df.LanguageWorkedWith)
```

```python
languages_worked_percent = languages_worked_df.mean().sort_values(ascending=False)

plt.figure(figsize=(12,12))
sns.barplot(x=languages_worked_percent*100, y=languages_worked_percent.index);
plt.xlabel("Percentage");
```

```python
# In which countries do developers work the highest number of hours per week?
countries_df = survey_df.groupby("Country")[["WorkWeekHrs", "Age"]].mean().sort_values("WorkWeekHrs", ascending=False)
# Counting countries with more than 250 responses only
high_response_countries = countries_df.loc[survey_df.Country.value_counts() > 250]
```
